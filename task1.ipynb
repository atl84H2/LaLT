### Задача 1.1 Расчет положения уровня Ферми

Для решения задачи будет оценено положение уровня Ферми относительно зоны проводимости $E_F-E_C$ и валентной зоны $E_V-E_F$

Импорт необходимых библиотек

```python
import numpy as np
import matplotlib.pyplot as plt
```

Параметры

```python
T=300 # температура в К
k_B=8.61*10**5 # постоянная Больцмана в эВ/К
m_0=9.1*10**31 # масса электрона в кг
m_C=0.014*m_0 # эффективная масса электрона в кг
m_V=0.43*m_0 # эффективная масса дырки в кг
```

Расчет

```python
n=np.linspace(10**21, 10**25, 10**4) # варьируем концентрацию в обратных метрах
N_C=2.5*10**19*(m_C/m_0)**1.5*(T/300)**1.5 # эффективная плотность состояний электронов
N_V=2.5*10**19*(m_V/m_0)**1.5*(T/300)**1.5 # эффективная плотность состояний дырок
```

```python
E_C=-k_B*T*np.log(N_C/n) # положение уровня Ферми относительно зоны проводимости
E_V=k_B*T*np.log(N_V/n) # положение уровня Ферми относительно зоны валентности
```

График

```python
%matplotlib widget
plt.figure(1)
plt.title('Зависимость положения уровня Ферми от концентрации носителей')
plt.xlabel('n и p, м$^{-3}$')
plt.ylabel('Положение уровня Ферми, эВ')
plt.plot(n, E_C, color='blue', label='электронов')
plt.plot(n, E_V, color='green', label='дырок')
plt.legend(loc='best')
plt.xscale('log')
plt.grid()
plt.savefig('graph1.1.png',dpi=600)
plt.show()
```

Логарифмическая шкала по оси абсцисс введена для наглядности экспоненциального характера зависимости. Но в основном продиктовано такое решение недомолвками с matplotlib, готов продемонстрировать на защите.
