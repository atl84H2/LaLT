### Задача 2.2 Определение уровня собственных потерь в резонаторе

Параметры

```python
R_m2=0.36
L=np.linspace(800, 1800, 1000)
```

Расчет

```python
alpha_m=1/(2*L)*np.log(1/R_m2)
```

График

```python
%matplotlib widget
plt.figure(1)
plt.title('Зависимость коэффициента собственных потерь от длины резонатора')
plt.xlabel('Длина резонатора, мкм')
plt.ylabel('Уровень собственных потерь, 1/мкм')
plt.plot(L, alpha_m)
plt.grid()
plt.savefig('graph2.2.png',dpi=600)
plt.show()
```

### Задача 2.4 Исследование полупроводниковой гетероструктуры как распределенного брэгговского отражателя

Введем показатели преломления GaAs и AlGaAs, толщины слоев и массив для исследования зависимости от количества слоев

```python
n_1, n_2 = 3.3, 2.9
h_GaAs, h_AlGaAs = 757*10**(-10), 862*10**(-10)
N=np.arange(0,50)
```

Расчет

```python
R_N=((1-(n_2/n_1)**(2*N))/(1+(n_2/n_1)**(2*N)))**2
```

График

```python
%matplotlib widget
plt.figure(1)
plt.title('Зависимость коэффициента отражения от количества слоев')
plt.xlabel('Число слоев, шт.')
plt.ylabel('Коэффициент отражения')
plt.plot(N, R_N)
plt.grid()
plt.savefig('graph2.4.png',dpi=600)
plt.show()
```

Определим общую толщину h (результат приведен в мкм)

```python
for i in range(len(R_N)):
    if R_N[i]>=0.98:
        h=i*(h_GaAs+h_AlGaAs)
        break
h*10**6
```

И рассчитаем время для синтеза

```python
v=1
t=h*10**6/v
t
```

Так, время для синтеза составляет около 3 часов 24 минут
